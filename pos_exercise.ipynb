```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
with open("news_story.txt","r") as f:
    news_text = f.read()

news_text[:500]
```

```python
doc = nlp(news_text)

numeral_tokens = []
noun_tokens = []

for token in doc:
    if token.pos_== 'NOUN':
        noun_tokens.append(token)
    elif token.pos_=='NUM':
        numeral_tokens.append(token)

numeral_tokens[:10]
```
